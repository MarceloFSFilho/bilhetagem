```python
from datetime import date, timedelta, datetime
from pathlib import Path
import time

caminho = Path('C:/Users/marcelo.francisco/Desktop')
#caminho = Path('Y:/v3lan/IMPORTACAO/HG/UD_202310/UD_20231025/BKP - BILHETAGEM')

#tg = input() + '.txt'

codigo = {'000': 'Indefinido', 
         '001': 'Vinculo Motorista', 
         '002': 'Vinculo Cobrador',
         '003': 'Vinculo Duplo',
         '004': 'Vinculo Motorista',
         '005': 'Vinculo Cobrador',
         '006': 'Vinculo Duplo'}

empresas = {'006': 'Flores',
            '010': 'Real Rio',
            '017': 'Rio dOuro',
            '025': 'Mageli',
            '027': 'Beira Mar'}

status = {'0': 'Normal',
          '1': 'Forçado'}


def converte_juliana_bil002(data):
    curr_date = "31/12/2002"
    curr_date_temp = datetime.strptime(curr_date, "%d/%m/%Y")
    new_date = curr_date_temp + timedelta(days=int(data))
    format_date = new_date.strftime("%d/%m/%Y")
    return format_date

def convert_hora(hora):
    return time.strftime("%H:%M:%S", time.gmtime(int(hora)))


#f = open (caminho/tg,'r' )
f = open (caminho/'TG006.txt','r' )


conteudo = f.readlines()
for linha in conteudo:
    if linha[0:3] == '010':
        tipoRegistro = linha[:3]
        versaoRegistro = linha[3:6]
        funcOperador = linha[6:9]
        tsp = linha[9:12]
        codProdata = linha[12:22]
        matricula = linha[22:32]
        nrOnibus = linha[32:42]
        data = linha[42:47]
        hora = linha[47:52]
        codOperacao = linha[52:55]
        status = linha[55:56]
        assinatura = linha[56:60]
        
        
        print("==================================================")
        print("Data = " + converte_juliana_bil002(data))
        print("Empresa = " + empresas[tsp])
        print("Codigo Prodata = " + codProdata)
        print("Matricula = " + matricula[5:])
        print("Carro = " + nrOnibus [6:])
        print("Data = "+ converte_juliana_bil002(data))
        print("Hora = " + convert_hora(hora))
        print("Codigo operação = " + codigo[codOperacao])
        print("Status = " + status)
        print("assinatura = " + assinatura)
```
